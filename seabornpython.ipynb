```python
import seaborn as sns
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri.head()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri)
plt.show()
```

```python
#seaborn barplot
```

```python
sns.set(style="whitegrid")
```

```python
iris=pd.read_csv('IRIS.csv')
```

```python
iris.head()
```

```python
iris.iloc[0:20,0:5]
```

```python
sns.barplot(x='species',y='sepal_length',data=iris)
plt.show()
```

```python
sns.barplot(x='species',y='sepal_length',hue="sepal_width",data=iris)
plt.show()
```

```python
sns.barplot(x='species',y='sepal_length',data=iris,palette='Blues_d')
plt.show()
```

```python
sns.barplot(x='species',y='sepal_length',data=iris,palette='rocket')
plt.show()
```

```python
sns.barplot(x='species',y='sepal_length',data=iris,palette='vlag')
plt.show()
```

```python
sns.barplot(x='species',y='sepal_length',data=iris,color='orange')
plt.show()
```

```python
#scatter plot
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris,color='red')
plt.show()
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris,hue='species',style='species')
plt.show()
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris,hue='petal_length')
plt.show()
```

```python
diamond=pd.read_csv('diamonds.csv')
```

```python
diamond.head()
```

```python
sns.distplot(diamond['price'],color='g')
plt.show()
```

```python
sns.distplot(diamond['price'],hist=False)
plt.show()
```

```python
sns.distplot(diamond['price'],bins=10,kde=False)
plt.show()
```

```python
sns.distplot(diamond['price'],vertical=True)
plt.show()
```

```python
sns.distplot(diamond['price'],bins=10,kde=False,vertical=True)
plt.show()
```

```python
sns.boxplot(x='species',y='sepal_length',data=iris)
plt.show()
```

```python
sns.boxplot(x='species',y='sepal_length',data=iris,palette="Set1")
plt.show()
```

```python
sns.boxplot(x='petal_length',y='petal_width',data=iris,hue='species')
plt.show()
```
